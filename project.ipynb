# Normalization Function

```python
import re

def clean_text(text):
    
    # convert to lowercase
    text = text.lower()
    
    # remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove HTML tags 
    text = re.sub(r'<.*?>', '', text)
    
    # remove special characters except letters and numbers
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
final_df['text'] = final_df['text'].apply(clean_text)
```

```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
train_encodings = tokenizer(
    X_train.tolist(),
    truncation=True,
    padding="max_length",
    max_length=128
)

test_encodings = tokenizer(
    X_test.tolist(),
    truncation=True,
    padding="max_length",
    max_length=128
)
```

# create dataset class

```python
!pip install torch
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class FakeNewsDataset(Dataset):

    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels.iloc[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

# create train and test dataset

```python
train_dataset = FakeNewsDataset(train_encodings, y_train)
test_dataset = FakeNewsDataset(test_encodings, y_test)
```

```python
print(len(train_dataset))
print(len(test_dataset))
```

# DataLoader

```python


train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)
```

```python
for batch in train_loader:
    print(batch.keys())
    break
```

# load the model

```python
from transformers import DistilBertForSequenceClassification

model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=2
)
```

# move model to device

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

If GPU is available, training will be much faster.

# Define Optimizer

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

# Training Loop

```python
from tqdm import tqdm

model.train()

for epoch in range(3):

    for batch in tqdm(train_loader):

        optimizer.zero_grad()

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

        loss = outputs.loss

        loss.backward()

        optimizer.step()

    print("Epoch completed")
```

# Evaluate model

```python
model.eval()

correct = 0
total = 0

with torch.no_grad():

    for batch in test_loader:

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)

        preds = torch.argmax(outputs.logits, dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

accuracy = correct / total
print("Test Accuracy:", accuracy)
```

# Prediction Function

```python
from transformers import DistilBertForSequenceClassification, DistilBertTokenizer
import torch

model = DistilBertForSequenceClassification.from_pretrained("bert_misinformation_model")
tokenizer = DistilBertTokenizer.from_pretrained("bert_misinformation_model")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

model.eval()
```

```python
def predict(text):

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=512
    )

    input_ids = inputs["input_ids"].to(device)
    attention_mask = inputs["attention_mask"].to(device)

    with torch.no_grad():

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

    logits = outputs.logits
    prediction = torch.argmax(logits, dim=1).item()

    return prediction
```

```python
import streamlit as st
import torch
import torch.nn.functional as F
from transformers import DistilBertForSequenceClassification, DistilBertTokenizer
from lime.lime_text import LimeTextExplainer

# Load model
model = DistilBertForSequenceClassification.from_pretrained("bert_misinformation_model")
tokenizer = DistilBertTokenizer.from_pretrained("bert_misinformation_model")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

labels = ["Fake", "Real"]

# Prediction function
def predict_proba(texts):

    inputs = tokenizer(
        texts,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=512
    )

    input_ids = inputs["input_ids"].to(device)
    attention_mask = inputs["attention_mask"].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)

    probs = F.softmax(outputs.logits, dim=1)

    return probs.cpu().numpy()


# Streamlit UI
st.title("Medical Misinformation Detection")

st.write("Enter a medical statement to check if it is misinformation.")

text = st.text_area("Enter Text")

if st.button("Predict"):

    if text:

        probs = predict_proba([text])[0]
        prediction = probs.argmax()

        confidence = probs[prediction]

        st.subheader("Prediction")
        st.write(labels[prediction])

        st.subheader("Confidence")
        st.write(f"{confidence*100:.2f}%")

        # Explainable AI using LIME
        st.subheader("Explanation")

        explainer = LimeTextExplainer(class_names=labels)

        exp = explainer.explain_instance(
            text,
            predict_proba,
            num_features=10
        )

        html = exp.as_html()
        st.components.v1.html(html, height=400, scrolling=True)
```
